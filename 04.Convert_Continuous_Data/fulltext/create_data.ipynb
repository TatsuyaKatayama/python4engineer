```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
np.random.seed(seed=100)

t = np.linspace(0,5.,5001)
t
```

```python
chan0 = np.random.rand() * np.sin(2*np.pi*t*10 - 2* np.pi * np.random.rand() ) + \
        np.random.rand() * np.sin(2*np.pi*t*20 - 2* np.pi * np.random.rand() ) + \
        np.random.rand() * np.sin(2*np.pi*t*30 - 2* np.pi * np.random.rand() ) + \
        np.random.rand() * np.sin(2*np.pi*t*40 - 2* np.pi * np.random.rand() ) + \
        np.random.rand() * np.sin(2*np.pi*t*50 - 2* np.pi * np.random.rand() ) 
chan0 -= chan0.min()
```

```python
noise = 0.5 * np.random.rand(5001)
```

```python
plt.plot(t,chan0+noise)
plt.xlim(1,1.5)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame([t,chan0+noise]).T
```

```python
df.columns = ["time(s)","channel_0"]
```

```python
df.plot(x=df.columns[0],xlim=(0,0.5))
```

```python
df.to_csv("sensor_data.csv",index=False)
```
